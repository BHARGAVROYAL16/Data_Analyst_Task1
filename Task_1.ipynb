```python
#Loading the Dataset
import pandas as pd
df = pd.read_csv('/content/market_sales_data_dirty.csv')
df.head()
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Fill missing Total Revenue as Units Sold * Unit Price
df['Total Revenue'] = df['Total Revenue'].fillna(df['Units Sold'] * df['Unit Price'])
```

```python
#Removing Duplicates
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

df = df.drop_duplicates()
```

```python
#Standardizing Text Values
# Example: lowercasing and stripping spaces for object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.lower().str.strip()
```
